# BORG SDSS data products
## borg_sdss_density package
* Authors: Jens Jasche, Florent Leclercq, Benjamin Wandelt
* Last update: 09/10/2018

This package contains some of the results obtained by [Jasche et al. (2015)](https://arxiv.org/abs/1409.6308), who performed a fully Bayesian analysis of the 3D large-scale structure as traced by SDSS main sample galaxies.

In particular, the package contains the ensemble mean of the density contrast for the observed present day large scale structure in terms of a 3D volumetric data cube. Corresponding uncertainty quantification is provided in terms of voxel-wise standard deviations.

Data are provided in terms of a standard 3D numpy array and can easily be accessed and processed using python.

For further details on the data and employed inference methodologies please consult [Jasche et al. (2015)](https://arxiv.org/abs/1409.6308) (reference below).

## File Content

This package consists of two files:
* **borg_sdss_density.ipynb**: The file you are currently reading.
* **borg_sdss_density.tar.gz**: Archive available on [GitHub LFS](https://github.com/florent-leclercq/borg_sdss_data_release/blob/master/borg_sdss_density/borg_sdss_density.tar.gz) and on [the author's website](http://www.florent-leclercq.eu/data/borg_sdss_density.tar.gz).

The archive borg_sdss_density.tar.gz can be extracted using

```bash
tar -xzf borg_sdss_density.tar.gz
```

It contains the file **borg_sdss_density.npz**.

## Usage

### Loading and accessing the data

The file **borg_sdss_density.npz** contains inferred large-scale structure data in a standard uncompressed .npz format. To load and process the data execute the following commands:

````python
import numpy as np
data=np.load('borg_sdss_density.npz')
````

To access the 3D ensemble mean final density contrast use: 

````python
#3D ensemble mean for the final density contrast 
mean=data['mean']
````

Individual voxels in this 3D volumetric data cube can be accessed as follows:

````python
k=10;j=127;i=243
delta_ijk=mean[k,j,i]
````

where i,j and k index voxel positions along the x,y and z axes respectively. All indices run from 0 to 255.

Similarly, to access voxel-wise standard deviations, use: 

````python
#3D pixelwise standard deviation for the final density contrast
stdv=data['stdv']
````

Individual voxels of this 3D volumetric data cube can be accessed as described above.

The ranges describing the extent of the cubic cartesian volume along the x,y and z axes can be accessed as follows:

````python
#Minimum and maximum position along the x-axis in Mpc/h
xmin=data['ranges'][0]
xmax=data['ranges'][1]

#Minimum and maximum position along the y-axis in Mpc/h
ymin=data['ranges'][2]
ymax=data['ranges'][3]

#Minimum and maximum position along the z-axis in Mpc/h
zmin=data['ranges'][4]
zmax=data['ranges'][5]
````

Units are Mpc/h.

(Note that all the maps that are part of the BORG SDSS data products have consistent coordinate systems. The coordinate transform to change from Cartesian to spherical coordinates and vice versa is given in appendix B of [Jasche et al. 2015](https://arxiv.org/abs/1409.6308)).

### Example plot

````python
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
%matplotlib inline
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,5))
im1 = ax1.imshow(mean[:,:,128], origin='lower', extent=[ymin,ymax,zmin,zmax], cmap="magma")
ax1.set_title("Density contrast: mean")
divider = make_axes_locatable(ax1)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = f.colorbar(im1, cax=cax)
im2 = ax2.imshow(stdv[:,:,128], origin='lower', extent=[ymin,ymax,zmin,zmax], cmap="Blues")
ax2.set_title("Density contrast: standard deviation")
divider = make_axes_locatable(ax2)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = f.colorbar(im2, cax=cax)
plt.show()
````
